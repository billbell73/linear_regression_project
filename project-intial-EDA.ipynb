```python
import pickle
import pandas as pd
import seaborn as sns
```

```python
with open('cars.pickle', 'rb') as handle:
    cars = pickle.load(handle)

len(cars)
```

```python
df = pd.DataFrame.from_dict(cars, orient='index')
df.head()
```

```python
cars_df = df.dropna()
cars_df.info()
```

```python
cars_df.describe()
```

```python
cars_df.sort_values("price", ascending=False).head(15)
```

```python
cars_df.sort_values("price").head(15)
```

```python
cars_df.value_counts(subset=['body_type'])
```

```python
cars_df['convertible'] = (cars_df['body_type']=='Convertible').astype(int)
cars_df.head()
```

```python
cars_df.value_counts(subset=['gearbox'])
```

```python
cars_df['manual'] = (cars_df['gearbox']=='Manual').astype(int)
cars_df.sample(5)
```

```python
cars_df.value_counts(subset=['fuel_type'])
```

```python
cars_df["years_old"] = 2021 - pd.to_numeric(cars_df["year"])
cars_df.head()
```

```python
numerics_df = cars_df.select_dtypes('number').astype(float)
numerics_df
```

```python
sns.pairplot(numerics_df)
```

```python
sns.heatmap(numerics_df.drop(columns=["doors", "cylinders"]).corr(), cmap=sns.diverging_palette(220, 5, n=20), annot=True, vmin=-1, vmax=1)
```
