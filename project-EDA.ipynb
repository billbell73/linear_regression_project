```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
ax = sns.regplot(x="power", y="price", data=df, truncate = True, line_kws = {'color': 'red'})

ax.set_xlabel('Engine power in bhp')
ax.set_ylabel('Listing price in GBP')
ax.set_title('Listing price of used porsche cars predicted by engine brake horsepower')

plt.savefig('porsches_price_from_power.png', dpi=300)
```

```python
sns.heatmap(numerics_df.drop(columns=["doors", "cylinders"]).corr(), cmap=sns.diverging_palette(220, 5, n=20, center="dark"), annot=True, vmin=-1, vmax=1);
```

```python
import statsmodels.api as sm
```

```python
X = numerics_df.drop(columns=["doors", "cylinders", "price"]).astype(float)
y = numerics_df.loc[:, 'price']
y.head()
```

```python
porsche_model = sm.OLS(y, X, data=numerics_df)
results = porsche_model.fit()
results.summary()
```

```python
model_with_const = sm.OLS(y, sm.add_constant(X), data=numerics_df)
results_with_const = model_with_const.fit()
results_with_const.summary()
```

```python
import matplotlib.pyplot as plt #define figure size
fig = plt.figure(figsize=(12,8))

#produce regression plots
# fig = sm.graphics.plot_regress_exog(results, 'price', fig=fig)

sm.graphics.plot_regress_exog(results, 'years_old', fig=fig)
plt.show()
```

```python
fig = plt.scatter(results.predict(), results.resid)
fig.axes.set_ylabel('Residual')
fig.axes.set_xlabel('Prediction')
```
